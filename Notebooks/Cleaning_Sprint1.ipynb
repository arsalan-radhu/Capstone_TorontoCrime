The data looks to have concatenated well and the number of columns have risen to <b>44!</b> So, I will be dropping some columns which contained either useless data, redundant data or data which are not necessary for my analysis. 

```python
# Doesn't make sense
final_df.drop('F_num_removed', axis=1, inplace=True)
final_df.drop('OCC_TIME_RANGE', axis=1, inplace=True)

# Old redundant data 
final_df.drop('NEIGHBOURHOOD_140', axis=1, inplace=True)
final_df.drop('HOOD_140', axis=1, inplace=True)
final_df.drop('REPORT_DATE', axis=1, inplace=True)
final_df.drop('OCC_DATE', axis=1, inplace=True)

# Resetting index since it got mixed up while concatanating 
final_df.reset_index(drop=True)

# Are not going to be used in my analysis
final_df.drop('BIKE_MAKE', axis=1, inplace=True)
final_df.drop('BIKE_MODEL', axis=1, inplace=True)
final_df.drop('BIKE_TYPE', axis=1, inplace=True)
final_df.drop('BIKE_SPEED', axis=1, inplace=True)
final_df.drop('BIKE_COLOUR', axis=1, inplace=True)
final_df.drop('BIKE_COST', axis=1, inplace=True)
```

```python
# Getting information about the data frame
final_df.info()
```

As we can see, after dropping all the unnecessary columns we are left with <b>31</b> columns. We can also observe that there seems to be a lot of `NULL` values. So, let's check for them!

```python
# Checking for the sum of all the NULL values
final_df.isna().sum()
```

As we can see there are a lot of `NULL` values. After doing some research I realized the `DEATH` and `INJURIES` columns were specific for `Shooting` dataset, and hence the huge amounts of `NULL` values. So, to fix that I will be filling the `NULL` values in these two columns with `0` as this will indicate that no-one was injured or killed in any other types of crimes.

```python
# Filling the NULL values
final_df["DEATH"] = final_df["DEATH"].fillna(0)
final_df["INJURIES"] = final_df["INJURIES"].fillna(0)
```

 The `HOMICIDE_TYPE` column is also specific to the `Homicide` dataset hence I will be filling the `NULL` values with "N/A". Similarly, `STATUS` and `PRIMARY_OFFENCE` is specific to the `Bike_Theft` data set, so I will be filling the `NULL` values with "N/A". The reason I am not dropping these columns is that even though they are specific to one type of crime, they provide very valuable insight into that crime, so for now I don't have any strong reasons to drop them.


```python
final_df["HOMICIDE_TYPE"] = final_df["HOMICIDE_TYPE"].fillna("NotApplicable")
final_df["STATUS"] = final_df["STATUS"].fillna("NotApplicable")
final_df["PRIMARY_OFFENCE"] = final_df["PRIMARY_OFFENCE"].fillna("NotApplicable")
```

`OFFENCE`, `UCR_EXIT`and `UCR_CODE` columns have exact same number of `NULL`, which implies that all of them might come from the same data set and after further investigation, that was the case. They were `NULL` due to them not being present in the `Bike Theft` data set, since they're not a required category for them we can fill these with `N/A` as well.


```python
final_df["MCI_CATEGORY"] = final_df["MCI_CATEGORY"].fillna("NonMCI")
final_df["OFFENCE"] = final_df["OFFENCE"].fillna("NotApplicable")
final_df["UCR_EXT"] = final_df["UCR_EXT"].fillna("NotApplicable")
final_df["UCR_CODE"] = final_df["UCR_CODE"].fillna("NotApplicable")
final_df["LOCATION_TYPE"] = final_df["LOCATION_TYPE"].fillna("NotApplicable")
final_df["PREMISES_TYPE"] = final_df["PREMISES_TYPE"].fillna("NotApplicable")
```

`REPORT_DATE_FIXED` has so many `NULL` values as `Homicide` and `Shootings` did not have any report dates or information. I am assuming that is the case as `Homicide` and `Shootings` both are severe crimes which usually don't have a difference in Occurence and reporting times. So we can fill some of the `NULL` values with the `OCC_DATE_FIXED`.

```python
final_df["REPORT_DATE_FIXED"] = final_df["REPORT_DATE_FIXED"].fillna(final_df["OCC_DATE_FIXED"])
```
